# How to open data files in pandas
You might have your data in .csv files or SQL tables. Maybe Excel files. Or .tsv files. 
But the goal is the same in all cases. If you want to analyze that data using pandas, the first step will be to read it into a data structure that’s compatible with pandas.

# Pandas data structures
There are two types of data structures in pandas: Series and DataFrames.

Series: a pandas Series is a one dimensional data structure (“a one dimensional ndarray”) that can store values — and for every value it holds a unique index, too.

DataFrame: a pandas DataFrame is a two (or more) dimensional data structure – basically a table with rows and columns. The columns have names and the rows have indexes.

```python
pd.read_csv("train.csv") 
```

```python
# makes the passed rows header 
pd.read_csv("train.csv", header =[1, 2]) 
```

```python
# make the passed column as index instead of 0, 1, 2, 3.... 
pd.read_csv("train.csv", index_col ='image_id') 
```

```python
# uses passed cols only for data frame 
pd.read_csv("train.csv", usecols =["label"]) 
```

```python
# returns pandas series if there is only one colunmn 
pd.read_csv("train.csv", usecols =["label"], squeeze = True) 
```

```python
# skips the passed rows in new series 
pd.read_csv("train.csv",  skiprows = [1, 2, 3, 4]) 
```
